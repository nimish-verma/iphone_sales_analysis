### top 10 highest rated iphones on flipkart in india


```python
data.head()
```

```python
# use the columns product name and star rating
highest_rated = data.sort_values(by = ["Star Rating"], ascending = False)
#to sort out the values in the column star_rating so that highest order are given either at the veginning of the dataset or the end

highest_rated_10 = highest_rated.head(10)
# print(highest_rated_10.head())
print(highest_rated_10["Product Name"], ["Star Rating"])
```

```python
print(highest_rated_10)
```

# ***How many Ratings on the highest rated iphones are present ?***

```python
data.head()
```

```python
highest_rated_10
```

```python
iphones = highest_rated_10['Product Name'].value_counts()
labels = iphones.index
counts = highest_rated_10["Number Of Ratings"]
figure = px.bar(highest_rated_10, x= labels, y= counts, title= "Number of Ratings on Highest Rated Iphones")
figure.show()
```

# ***Highest no of reviews on flipkart***

```python
data.head()
```

```python
iphones = highest_rated_10['Product Name'].value_counts()
labels = iphones.index
counts = highest_rated_10["Number Of Reviews"]
figure = px.bar(highest_rated_10, x= labels, y= counts, title= "Number of Reviews")
figure.show()
```

# ***Relationship b/w sale prices of iphone and the no of ratings on flipkart***

```python
figure = px.scatter(data_frame = data, x = "Number Of Ratings", y = "Sale Price", size = "Discount Percentage", trendline = "ols", title = "Relationship b/w Sale Price and Number of Ratings")
figure.show()
```

# ***Relationship b/w discount percentage and the no of ratings on flipkart***

```python
figure = px.scatter(data_frame = data,
                    x="Number Of Ratings",
                    y = "Discount Percentage",
                    trendline="ols",
                    size = "Sale Price",
                    title = "Discount and Ratings"
                    )
figure.show()
```
